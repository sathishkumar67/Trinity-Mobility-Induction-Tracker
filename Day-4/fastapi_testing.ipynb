```python
import requests

url = "http://127.0.0.1:8000/users"
params = {"api_key": "secret123"}
data = {"name": "Alex", "age": 25}

requests.post(url, params=params, json=data).json()
```

```python
requests.get("http://127.0.0.1:8000/users").json()
```

```python
requests.get("http://127.0.0.1:8000/users/0").json()
```

```python
requests.delete(
    "http://127.0.0.1:8000/users/0",
    params={"api_key": "secret123"}
).json()
```
